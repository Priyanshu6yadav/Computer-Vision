```python
import cv2
```

```python
## Reading a image
image = cv2.imread("pexels-pixabay-459203.jpg")
```

```python
# check if image is load successfully
if image is None:
    print("image is not lodading")
else:
    print("Image is successfully load")
```

```python
cv2.imshow("Image window", image) ## Open a windows and show image
cv2.waitKey(0) # pause the program untill a key is press
cv2.destroyAllWindows() # close all open windows
```

```python
# Write the image
image_new = image + 30 ## add 30 in all of the pixels
cv2.imwrite("./output.jpg", image_new)
```
