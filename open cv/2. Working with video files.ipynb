```python
import cv2
```

```python
## Load the video
video = cv2.VideoCapture("Mountain_Valley_Nature_Video_Creation.mp4")
```

```python
## Continuously read and display video frames
while True:
    # Read one frame from the video
    # ret → True if frame is successfully read, False if video ends or fails
    # frame → the actual image array of the current frame
    ret, frame = video.read()
    
    # If no frame is returned (end of video), stop the loop
    if not ret:
        break

    # Display the current video frame in a window named "video frame"
    cv2.imshow("video frame", frame)

    # Wait for 1 millisecond for a key press
    # cv2.waitKey(1) returns a 32-bit integer
    # & 0xFF keeps only the last 8 bits (ASCII value)
    # ord('q') gets ASCII value of character 'q'
    # If 'q' is pressed, exit the loop
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break


# Release the video file or camera resource
# This is important to free memory and hardware access
# Close all OpenCV windows
video.release()
cv2.destroyAllWindows()
cv2.waitKey(1)   # FORCE event processing
```

```python
import cv2

# Open the input video file for reading
video = cv2.VideoCapture("Mountain_Valley_Nature_Video_Creation.mp4")

# Define the codec to write the output video (mp4 format)
# *'mp4v' expands to four characters: 'm', 'p', '4', 'v'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Get the width of the video frames
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))

# Get the height of the video frames
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Create a VideoWriter object to save the output video
# Parameters:
# filename, codec, frame rate (FPS), frame size (width, height)
out = cv2.VideoWriter('output.mp4', fourcc, 30, (width, height))

# Read and write video frames in a loop
while True:
    # Read one frame from the input video
    # ret → True if frame is read successfully, False if video ends
    # frame → the current video frame
    ret, frame = video.read()

    # If no frame is returned, stop processing
    if not ret:
        break

    # Write the current frame to the output video file
    out.write(frame)

# Release the VideoWriter object and save the output file
out.release()

# Release the input video resource
video.release()

# Close any OpenCV windows (safe even if none are open)
cv2.destroyAllWindows()
```

```python
# Write a video
video = cv2.VideoCapture("Mountain_Valley_Nature_Video_Creation.mp4")
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
hight = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter('output_new.mp4', fourcc, 30, (width,hight))

while True:
    ret, frame = video.read()
    if not ret:
        break
    frame = cv2.add(frame, 20)   # brighten frame
    out.write(frame)

out.release()
video.release()
cv2.destroyAllWindows()
```
